# Read excel

```python
labels = pd.read_excel("population salary analysis.xlsx", sheet_name='Salary band', usecols=['Salary Band'])
# labels = labels[['Salary Band']]
labels = labels['Salary Band'].tolist()
labels
```

```python
bins = [0, 200_000, 400_000, 600_000, 800_000, 1000_000,
        1200_000, 1400_000, 1600_000, 1800_000, np.inf]

df['salary_band'] = pd.cut(df['salary'], bins=bins, labels=labels)
```

```python
df.head()
```
